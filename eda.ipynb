```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# STEP 1: load the dataset
data = pd.read_csv("datasets/supermarket_data.csv")

# STEP 2: plot
sns.countplot(x="gender", data=data, palette="spring")

plt.title("Gender Distribution Against Total Count")
plt.ylabel("Total Count of Individuals")
plt.show()
```

```python
# Class Work: Create a Count Plot of product
# Which product category appears the most?
sns.countplot(x="product", data=data, palette="winter")
plt.xticks( rotation = 45)
```

```python
# Below is a countplot that has two coloumns
sns.countplot(x="gender", hue="city", palette="Reds", data=data)
```
